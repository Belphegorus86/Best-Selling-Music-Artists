```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import scipy.stats as stats


df = pd.read_csv("https://raw.githubusercontent.com/Belphegorus86/Best-Selling-Music-Artists-of-the-70-s-80-s-and-90-s/master/Best-Selling-Music-Artists-of-the-70-s-80-s-and-90-s/CSV/best_selling_artists.csv")
df
```

```python
df.info() 
```

```python
df.columns = df.columns.str.lower().str.replace('tcu', 'units sold').str.replace('_', ' ').str.replace('units sold', 'units sold in millions').str.replace('sales', 'sales in millions').str.replace('year', '1st time on top charts')
df
```

```python
#Droped column "period active", dont think its really initeresting to keep
df = df.drop('period active', axis=1)
df
```

```python
# So here I changed the word "million" to " ", i wanted to keep add numbers instead of just having so few numbers 
# But this would create an issue with "units sold in millions" because if we add "000000" instead of just " " the lines
# the lines that have a "." they would have and extra digit
# I think we can solve this by using When-if Statements, but im not sure
#if you find a way please let me know
df = df.apply(lambda x: x.replace({' million':' '}, regex=True))
df
```

```python
#Transformed the columns "units sold in millions " "and sales in millions " from object to float 
df['units sold in millions'] = df['units sold in millions'].astype(float)
df['sales in millions'] = df['sales in millions'].astype(float)
df.info() 
```
